```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Untuk menghasilkan data latih
np.random.seed(42)
x_train = np.random.rand(100, 1) * 10
y_train = 2 * x_train + 2 + np.random.randn(100, 1)
```

```python
# Membangun model jaringan saraf
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[1], activation='linear')
])

# Mengompilasi model
model.compile(optimizer='adam', loss='mean_squared_error')

# Melatih model
model.fit(x_train, y_train, epochs=1000, verbose=0)
```

```python
# Testing dengan data baru
x_test = np.array([[5.0]])
y_pred = model.predict(x_test)
print(f"Predicted y for x=5: {y_pred[0, 0]}")
```

```python
# Plot the training data and the learned function
plt.scatter(x_train, y_train, label='Training Data')
x_range = np.linspace(0, 10, 100).reshape(-1, 1)
y_range = model.predict(x_range)
plt.plot(x_range, y_range, color='red', label='Learned Function (2x + 2)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Lists to store training history
loss_history = []
prediction_history = []
```

```python
# Train the model and store history
for epoch in range(1000):
    history = model.fit(x_train, y_train, epochs=1, verbose=0)
    loss_history.append(history.history['loss'][0])

    # Generate predictions every 100 epochs
    if epoch % 100 == 0:
        y_pred = model.predict(x_train)
        prediction_history.append(y_pred)
```

```python
# Plot the training data and the learned function
plt.figure(figsize=(12, 4))
```

```python
plt.subplot(1, 2, 2)
plt.plot(range(1, len(loss_history) + 1), loss_history, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
